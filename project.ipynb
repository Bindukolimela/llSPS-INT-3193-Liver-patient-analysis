# Encoding Categorical Columns using Label Encoder :-

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
x[:,1]=lb.fit_transform(x[:,1])
x
```

```python
y=lb.fit_transform(y)
y
```

# Feature Scaling :-

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from joblib import dump
dump(sc,"analysis.save")
```

# Splitting Train and Test Data :-

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

The given Dataset shows the output with highest Accuracy in Logistic Regression. so, We use Logistic Regression

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
import pickle
pickle.dump(lr,open("LR.pkl","wb"))
```

```python
y_pa=lr.predict(x_test)
y_pa
```

```python
y_test
```

```python
lr.predict([[70,0,10.9,0.1,699,70,68,7.0,6.6,190]])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pa)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pa)
cm
```

```python
import sklearn.metrics as metrics
fpr,tpr,threshold=metrics.roc_curve(y_test,y_pa)
roc_auc=metrics.auc(fpr,tpr)
roc_auc
```

```python
plt.title("receiveroperatingcharacteristics")
plt.plot(fpr,tpr,"b",label="AUC=%0.2f")
plt.legend(loc="lower right")
plt.plot
```
